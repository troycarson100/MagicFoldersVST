```python
# If your data is in a zip: upload it to Colab, then run:
# !unzip -q -o training_data.zip
# Then set DATA_DIR to the folder that contains Kick/, Snare/, etc.
# Use the full path so it works no matter where we run the script from:
DATA_DIR = "/content/MagicFoldersTraining"  # or "/content/training_data" or "/content/data" — whatever your unzipped folder is called
```

```python
# Run from the folder that contains train_and_export.py AND model_constants.py so the import works
import os
training_dir = None
for candidate in ["MagicFolders-main/training", "MagicFoldersVST/training", "training", "."]:
    path = os.path.join("/content", candidate) if not os.path.isabs(candidate) else candidate
    if os.path.isfile(os.path.join(path, "train_and_export.py")) and os.path.isfile(os.path.join(path, "model_constants.py")):
        training_dir = os.path.abspath(path)
        break
if training_dir is None:
    raise FileNotFoundError("Could not find folder with train_and_export.py and model_constants.py. Unzip the Magic Folders repo so you have MagicFolders-main/training/ (or similar).")
os.chdir(training_dir)
print("Working directory:", os.getcwd())
!python train_and_export.py --data_dir "$DATA_DIR" --epochs 30 --out_model InstrumentClassifier.onnx
```
